```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
file_path = "data/SWE.pdf"
loader = PyPDFLoader(file_path)
data = loader.load()
```

```python
len(data) # pages
```

```python
question_gen = ""
for page in data:
    question_gen += page.page_content
```

```python
question_gen
```

```python
from langchain_text_splitters import TokenTextSplitter
```

```python
splitter_ques_gen = TokenTextSplitter(
    model_name= "gpt-4.1-mini",
    chunk_size= 10000,
    chunk_overlap = 200
)
```

```python
chunk_ques_gen = splitter_ques_gen.split_text(question_gen)
```

```python
chunk_ques_gen
```

```python
len(chunk_ques_gen)
```

```python
type(chunk_ques_gen[0])
```

```python
from langchain_core.documents import Document
```

```python
document_ques_gen = [Document(page_content=t) for t in chunk_ques_gen]
```

```python
document_ques_gen
```

```python
type(document_ques_gen[0])
```

```python
splitter_ans_gen = TokenTextSplitter(
    model_name= "gpt-4.1-mini",
    chunk_size= 256,
    chunk_overlap = 25
)
```

```python
document_ans_gen = splitter_ans_gen.split_documents(
    document_ques_gen
)
```

```python
document_ans_gen
```

```python
from langchain_openai import ChatOpenAI
```

```python
llm_ques_gen_pipeline = ChatOpenAI(
    model = 'gpt-4.1-mini',
    temperature = 0.3
)
```

```python
prompt_template = """
You are an expert at creating questions based on coding job description.
Your goal is to prepare a coder or programmer for their exam and coding tests.
You do this by asking questions based on the job description below:

------------
{text}
------------

Create questions that will prepare the coders or programmers for their tests.
Make sure not to lose any important information.

QUESTIONS:
"""
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
PROMPT_QUESTIONS = PromptTemplate(template=prompt_template, input_variables=['text'])
```

```python
refine_template = ("""
You are an expert at creating practice questions based on coding job description.
Your goal is to help a coder or programmer prepare for a coding test.
We have received some practice questions to a certain extent: {existing_answer}.
We have the option to refine the existing questions or add new ones.
(only if necessary) with some more context below.
------------
{text}
------------

Given the new context, refine the original questions in English.
If the context is not helpful, please provide the original questions.
QUESTIONS:
"""
)
```

```python
REFINE_PROMPT_QUESTIONS = PromptTemplate(
    input_variables=["existing_answer", "text"],
    template=refine_template,
)
```

```python
from operator import itemgetter
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

initial_chain = (
    {"text": RunnablePassthrough()} 
    | PROMPT_QUESTIONS 
    | llm_ques_gen_pipeline 
    | StrOutputParser()
)

ques_gen_chain = (
    {
        "existing_answer": initial_chain, 
        "text": RunnablePassthrough() 
    }
    | REFINE_PROMPT_QUESTIONS
    | llm_ques_gen_pipeline
    | StrOutputParser()
)
```

```python
ques = ques_gen_chain.invoke(document_ques_gen)
```

```python
ques
```

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embeddings = OpenAIEmbeddings()
```

```python
from langchain_community.vectorstores import FAISS
```

```python
vector_stores = FAISS.from_documents(document_ans_gen, embeddings)
```

```python
llm_answer_gen = ChatOpenAI(temperature=0.1, model="gpt-4.1-mini")
```

```python
ques_list = ques.split("\n")
```

```python
ques_list
```

```python
# Lọc: Chỉ giữ lại các dòng bắt đầu bằng số (câu hỏi)
filtered_questions = [line for line in ques_list if line.strip() and line.strip()[0].isdigit()]

# In ra xuống dòng cho đẹp
print("\n".join(filtered_questions))
```

```python
retriever = vector_stores.as_retriever()
```

```python
for question in filtered_questions:
    print("Question: ", question)
    answer = retriever.invoke(question)
    combined_answer = "\n".join([doc.page_content for doc in answer])
    print("Answer: ", combined_answer)
    print("------------------------------------------------------------\\n\\n")
    # Save answer to file
    with open("answers.txt", "a") as f:
        f.write("Question: " + question + "\\n")
        combined_answer = "\n".join([doc.page_content for doc in answer])
        f.write("Answer: " + combined_answer + "\\n")
        f.write("------------------------------------------------------------")
```
